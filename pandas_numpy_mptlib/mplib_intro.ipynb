```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red')
plt.title("Scatter Example")
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [4, 7, 1]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green')
plt.title("Histogram")
plt.show()
```

```python
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
plt.subplot(1, 2, 1)   # (rows, columns, index)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(x, -y)

plt.tight_layout()
plt.show()
```
